##1. Calculadora de l'√≠ndex de massa corporal

Escriu una funci√≥ que calculi l'IMC ingressat per l'usuari/√†ria, √©s a dir, qui ho executi haur√† d'ingressar aquestes dades. Pots obtenir m√©s informaci√≥ del seu c√†lcul en: üîó √çndice de masa corporal IMC que es y como se calcula.
La funci√≥ ha de classificar el resultat en les seves respectives categories.
Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introdu√Ødes per l‚Äôusuari/a no estan en el format adequat o no pren valors raonables.

```python
def imc():
    while True: #creamos un bucle hasta que se introduzcan datos correctos
        try:
            altura = float(input('Ingresa tu altura en metros (por ejemplo 1.65): '))
            peso = float(input('Ingresa tu peso en kilos (por ejemplo 55): '))
            break #detiene el while, ya tenemos los datos correctos
        except ValueError: # si se indroduce la altura o peso incorrectamente, eso lanza un ValueError y Python abandona inmediatamente el bloque try y salta al except ValueError, no llega a ninguna linea posterior, no llega el break
            print('Por favor, ingresa solo n√∫meros v√°lidos (usa punto decimal, no coma).')

    indice = peso / (altura ** 2)
    print('Tu IMC es:', round(indice, 2))

imc()
```

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

##4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client t√© una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al rev√©s, √©s a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original s√≥n √∫nics; si aquest no √©s el cas, la funci√≥ hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
def invertir_diccionario(dic):
    dic_inverso = {} # creamos el diccionario inverso
    duplicados = {} # creamos un diccionario para los duplicados

    for clave, valor in dic.items(): # iteramos el diccionario (clave:valor)
        if valor not in dic_inverso: # si el valor no est√° ya, lo ponemos en dic_invers (el diccionario no admite 2 claves iguales)
            dic_inverso[valor] = clave
        else:                       # si el valor ya est√° (duplicado)
            if valor not in duplicados:
                duplicados[valor] = [dic_inverso[valor]] # lo a√±adimos a duplicados ya que no est√°
            duplicados[valor].append(clave) # guardamos en una lista las claves que son iguales
            dic_inverso[valor] = duplicados[valor]  # reflejamos asi en dic_invers todas las claves que comparten el mismo valor

    # Mensaje de advertencia centralizado
    if duplicados: # solo se considera True si no es vacio
        print('Hemos encontrado valores duplicados:')
        for valor, claves in duplicados.items():
            print(f' {valor} - {claves}')

    return dic_inverso

# Ejemplo: lo substituiremos por el diccionario del cliente cuando nos lo de
enquesta = {'Ana': 7,'Dragos': 9,'Laura': 8,'Luis': 9}
resultado = invertir_diccionario(enquesta)
print('\nDiccionario Inverso:')
print(resultado)
```

##2. Conversi√≥ de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
def convertir_lista(lista):
    convertidos = []
    no_convertidos = []

    for element in lista:
        # si el elemento es una lista o tupla, procesamos recursivamente
        if isinstance(element, (list, tuple)):
            conv, no_conv = convertir_lista(element)
            convertidos.extend(conv)
            no_convertidos.extend(no_conv)
        else:
            # limpiamos espacios y convertimos comas a puntos
            valor = str(element).strip().replace(',', '.')

            try:
                convertidos.append(float(valor))
            except ValueError:
                no_convertidos.append(element)

    return convertidos, no_convertidos


# Ejemplo
data = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
convertidos, no_convertidos = convertir_lista(data)
print("Convertidos a float:", convertidos)
print("No convertidos:", no_convertidos)
```

##1. Generador de Contrasenyes

Explora el funcionament del m√≤dul random de la llibreria numpy. üîó Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d‚Äôaniversari o similars. Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

crear_contrasenya(10, True, True, True, True)

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).


```python
import numpy as np
import string
import pyperclip  # para copiar al portapapeles

def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False):
    # validamos que la longitud sea positiva
    if longitud <= 0:
        print("La longitud ha de ser un n√∫mero positivo.")
        return None

    caracters = ''
    opcions = []

    # a√±adimos los tipos de caracteres seleccionados
    if majuscules:
        opcions.append(list(string.ascii_uppercase))
        caracters += string.ascii_uppercase
    if minuscules:
        opcions.append(list(string.ascii_lowercase))
        caracters += string.ascii_lowercase
    if numeros:
        opcions.append(list(string.digits))
        caracters += string.digits
    if signes:
        opcions.append(list("!@#$%&*?,-$"))
        caracters += "!@#$%&*?,-$"

    # verificamos que haya al menos un tipo seleccionado
    if not caracters:
        print("Debes seleccionar al menos un tipo de car√°cter.")
        return None

    # nos aseguramos de que aparezca al menos un car√°cter de cada tipo seleccionado
    contrasenya = [np.random.choice(opcion) for opcion in opcions]

    # rellenamos hasta la longitud deseada con caracteres aleatorios
    if longitud > len(contrasenya):
        contrasenya += list(np.random.choice(list(caracters), longitud - len(contrasenya)))

    # mezclamos todo para que sea realmente aleatorio
    np.random.shuffle(contrasenya)

    # convertimos la lista en string
    contrasenya_final = ''.join(contrasenya[:longitud])  # nos aseguramos de la longitud exacta

    # copiamos autom√°ticamente al portapapeles
    try:
        pyperclip.copy(contrasenya_final)
        print("Contrasenya generada:", contrasenya_final)
        print("‚úì Contrase√±a copiada al portapapeles")
    except:
        print("Contrasenya generada:", contrasenya_final)
        print("(No se pudo copiar al portapapeles autom√°ticamente)")

    return contrasenya_final


# Ejemplo de uso
contrasenya = crear_contrasenya(10, True, True, True, True)
print("\nContrasenya guardada en variable:", contrasenya)
```

##2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. T√© un hist√≤ric de partits de futbol catal√† en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera autom√†tica, per extreure‚Äôn els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l‚Äôequip m√©s golejador.
El nom de l‚Äôequip m√©s golejat
La classificaci√≥ global (cada vict√≤ria: 3 pts, empat 1 pts, derrota 0 pts)
Com a refer√®ncia, els equips amb m√©s punts serien els seg√ºents:

imatge resultat partits
Pots utilitzar el format que et sigui m√©s convenient.

```python
def procesar_partidos(ruta_archivo):

    # Diccionario para almacenar estad√≠sticas de cada equipo
    equipos = {}

    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            for linea in archivo:
                linea = linea.strip()
                if not linea:  # Saltar l√≠neas vac√≠as
                    continue

                # Formato esperado: Equipo1 goles1-goles2 Equipo2
                # Ejemplo: Girona FC 3-1 Sabadell
                partes = linea.split()

                # Buscar el gui√≥n que separa los goles
                indice_resultado = -1
                for i, parte in enumerate(partes):
                    if '-' in parte:
                        indice_resultado = i
                        break

                if indice_resultado == -1:
                    continue  # L√≠nea mal formateada

                # Extraer nombres de equipos y resultado
                equipo1 = ' '.join(partes[:indice_resultado])
                resultado = partes[indice_resultado]
                equipo2 = ' '.join(partes[indice_resultado + 1:])

                # Extraer goles
                try:
                    goles1, goles2 = map(int, resultado.split('-'))
                except ValueError:
                    continue  # Resultado mal formateado

                # Inicializar equipos si no existen
                if equipo1 not in equipos:
                    equipos[equipo1] = {
                        'goles_favor': 0,
                        'goles_contra': 0,
                        'partidos_ganados': 0,
                        'partidos_empatados': 0,
                        'partidos_perdidos': 0,
                        'puntos': 0
                    }

                if equipo2 not in equipos:
                    equipos[equipo2] = {
                        'goles_favor': 0,
                        'goles_contra': 0,
                        'partidos_ganados': 0,
                        'partidos_empatados': 0,
                        'partidos_perdidos': 0,
                        'puntos': 0
                    }

                # Actualizar goles
                equipos[equipo1]['goles_favor'] += goles1
                equipos[equipo1]['goles_contra'] += goles2
                equipos[equipo2]['goles_favor'] += goles2
                equipos[equipo2]['goles_contra'] += goles1

                # Actualizar puntos y resultados
                if goles1 > goles2:  # Gana equipo1
                    equipos[equipo1]['partidos_ganados'] += 1
                    equipos[equipo1]['puntos'] += 3
                    equipos[equipo2]['partidos_perdidos'] += 1
                elif goles1 < goles2:  # Gana equipo2
                    equipos[equipo2]['partidos_ganados'] += 1
                    equipos[equipo2]['puntos'] += 3
                    equipos[equipo1]['partidos_perdidos'] += 1
                else:  # Empate
                    equipos[equipo1]['partidos_empatados'] += 1
                    equipos[equipo1]['puntos'] += 1
                    equipos[equipo2]['partidos_empatados'] += 1
                    equipos[equipo2]['puntos'] += 1

        return equipos

    except FileNotFoundError:
        print(f"Error: No se encontr√≥ el archivo '{ruta_archivo}'")
        return None
    except Exception as e:
        print(f"Error al procesar el archivo: {e}")
        return None


def mostrar_estadisticas(equipos):

    # muestra todas las estad√≠sticas calculadas
    if not equipos:
        print("No hay datos para mostrar.")
        return

    print("=" * 80)
    print("ESTAD√çSTICAS DE EQUIPOS - F√öTBOL CATAL√ÅN")
    print("=" * 80)
    print()

    # 1. Goles totales por equipo
    print("1. GOLES TOTALES POR EQUIPO:")
    print("-" * 50)
    goles_ordenados = sorted(equipos.items(),
                            key=lambda x: x[1]['goles_favor'],
                            reverse=True)

    for equipo, stats in goles_ordenados:
        print(f"   {equipo}: {stats['goles_favor']} goles")
    print()

    # 2. Equipo m√°s goleador
    equipo_mas_goleador = max(equipos.items(), key=lambda x: x[1]['goles_favor'])
    print("2. EQUIPO M√ÅS GOLEADOR:")
    print("-" * 50)
    print(f"   {equipo_mas_goleador[0]} con {equipo_mas_goleador[1]['goles_favor']} goles")
    print()

    # 3. Equipo m√°s goleado
    equipo_mas_goleado = max(equipos.items(), key=lambda x: x[1]['goles_contra'])
    print("3. EQUIPO M√ÅS GOLEADO:")
    print("-" * 50)
    print(f"   {equipo_mas_goleado[0]} con {equipo_mas_goleado[1]['goles_contra']} goles en contra")
    print()

    # 4. Clasificaci√≥n global
    print("4. CLASIFICACI√ìN GLOBAL:")
    print("-" * 50)
    clasificacion = sorted(equipos.items(),
                          key=lambda x: (x[1]['puntos'], x[1]['goles_favor']),
                          reverse=True)

    print(f"{'Pos':<5} {'Equipo':<25} {'PG':<5} {'PE':<5} {'PP':<5} {'GF':<5} {'GC':<5} {'Pts':<5}")
    print("-" * 80)

    for pos, (equipo, stats) in enumerate(clasificacion, 1):
        print(f"{pos:<5} {equipo:<25} {stats['partidos_ganados']:<5} "
              f"{stats['partidos_empatados']:<5} {stats['partidos_perdidos']:<5} "
              f"{stats['goles_favor']:<5} {stats['goles_contra']:<5} "
              f"{stats['puntos']:<5}")

    print("=" * 80)


# Programa principal
if __name__ == "__main__":
    # Ruta al archivo (ajusta seg√∫n tu ubicaci√≥n)
    ruta_archivo = "/content/historic partits.txt"

    print("Procesando hist√≥rico de partidos...")
    print()

    # Procesar el archivo
    equipos = procesar_partidos(ruta_archivo)

    # Mostrar estad√≠sticas
    if equipos:
        mostrar_estadisticas(equipos)
    else:
        print("No se pudieron procesar los datos.")
```
