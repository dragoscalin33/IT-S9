##1. Calculadora de l'√≠ndex de massa corporal

Escriu una funci√≥ que calculi l'IMC ingressat per l'usuari/√†ria, √©s a dir, qui ho executi haur√† d'ingressar aquestes dades. Pots obtenir m√©s informaci√≥ del seu c√†lcul en: üîó √çndice de masa corporal IMC que es y como se calcula.
La funci√≥ ha de classificar el resultat en les seves respectives categories.
Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introdu√Ødes per l‚Äôusuari/a no estan en el format adequat o no pren valors raonables.

```python
def imc():
    while True: #creamos un bucle hasta que se introduzcan datos correctos
        try:
            altura = float(input('Ingresa tu altura en metros (por ejemplo 1.65): '))
            peso = float(input('Ingresa tu peso en kilos (por ejemplo 55): '))
            break #detiene el while, ya tenemos los datos correctos
        except ValueError: # si se indroduce la altura o peso incorrectamente, eso lanza un ValueError y Python abandona inmediatamente el bloque try y salta al except ValueError, no llega a ninguna linea posterior, no llega el break
            print('Por favor, ingresa solo n√∫meros v√°lidos (usa punto decimal, no coma).')

    indice = peso / (altura ** 2)
    print('Tu IMC es:', round(indice, 2))

imc()
```

##2. Convertidor de temperatures

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les m√©s comunes s√≥n Celsius (¬∞C), Fahrenheit (¬∞F) i Kelvin (K). Tamb√© existeixen altres unitats com Rankine (¬∞Ra) i R√©aumur (¬∞Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a m√≠nim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades pr√®viament, perqu√® envi√Ø un missatge d‚Äôadvert√®ncia si les dades introduides per l‚Äôusuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d‚Äôemmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d‚Äôescriure molts if else en funci√≥ de la temperatura d‚Äôorigen i la temperatura de dest√≠.

```python
def convertidor():
    while True: # 1 bucle: hasta que se introduzca la temperatura correcta
        try:
            temperatura = float(input("Introduce la temperatura: "))
            break  # detiene el while, ya tenemos los datos correctos
        except ValueError:
            print('Por favor, ingresa solo n√∫meros v√°lidos (usa punto decimal, no coma).')

    while True: # 2 bucle: hasta que se introduzca C o F
        grados = input("Introduce C para Celsius o F para Fahrenheit: ").upper()
        if grados in ['C', 'F']:
            break  # si es correcto, sale del bucle
        else:
            print('Por favor, ingresa s√≥lo C o F.')

    # Conversiones
    if grados == 'C':
        celsius = temperatura
        fahrenheit = round((9/5 * celsius) + 32, 2)
        kelvin = round(celsius + 273.15, 2)
        rankine = round(9/5 * (celsius + 273.15), 2)
        temperaturas = (celsius, fahrenheit, kelvin, rankine)
        print('Las conversiones son en orden: Celsius, Fahrenheit, Kelvin, Rankine')

    elif grados == 'F':
        fahrenheit = temperatura
        celsius = round((fahrenheit - 32) * 5/9, 2)
        kelvin = round(celsius + 273.15, 2)
        rankine = round(fahrenheit + 459.67, 2)
        temperaturas = (celsius, fahrenheit, kelvin, rankine)
        print('Las conversiones son en orden: Celsius, Fahrenheit, Kelvin, Rankine')

    return temperaturas

convertidor()
```

##3. Comptador de paraules d'un text.

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

```python
import string

def contador():
  text = input('Introduce un texto, te dir√© cada vez que aparece cada palabra y cual es la media de la longitud de las palabras: ')
  text = text.lower()  # todo min√∫sculas

  for char in string.punctuation:  # string.punctuation es una cadena que contiene los simbolos mas comunes = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    text = text.replace(char, '')  # iteramos uno de estos simbolos a uno, y si lo encuentra en text, lo elimina (substituye por nada)

  palabras = text.split() # funcion que divide text en palabras y las pone en una lista
  contador_palabras = {} # definimos un diccionario. Perfecto para lo que queremos. Palabra:Numero de veces que aparece la palabra

  for palabra in palabras: # iteramos por la lista palabras y contamos cuantas veces aparece cada una y las ponemos en el diccionario
    if palabra in contador_palabras:
        contador_palabras[palabra] += 1 # como la palabra ya a aparedido antes, solo sumamos el valor +1
    else:
        contador_palabras[palabra] = 1 # aqui ponemos palabra por palabra (dupla clave valor) en el diccionario

  suma_longituds = sum(len(p) for p in palabras) # iteramos por todas las palabras y con len sabemos la longitud de cada palabra, las sumamos todas
  media = suma_longituds / len(palabras) # para calcular la media es solo dividir por el numero de palabras

  print('Contador de palabras:') # impriminos los resultados iterando por el diccionario
  for palabra, veces in contador_palabras.items(): # usamos .items() porque queremos los pares (clave, valor), .keys()solo para las claves y .values() solo para los valores

    print(f'{palabra}: {veces} veces')

  print(f'\nLongitud media de las palabras: {media:.2f}')

contador()
```

##4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client t√© una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al rev√©s, √©s a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original s√≥n √∫nics; si aquest no √©s el cas, la funci√≥ hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
def invertir_diccionario(dic):
    dic_inverso = {} # creamos el diccionario inverso
    duplicados = {} # creamos un diccionario para los duplicados

    for clave, valor in dic.items(): # iteramos el diccionario (clave:valor)
        if valor not in dic_inverso: # si el valor no est√° ya, lo ponemos en dic_invers (el diccionario no admite 2 claves iguales)
            dic_inverso[valor] = clave
        else:                       # si el valor ya est√° (duplicado)
            if valor not in duplicados:
                duplicados[valor] = [dic_inverso[valor]] # lo a√±adimos a duplicados ya que no est√°
            duplicados[valor].append(clave) # guardamos en una lista las claves que son iguales
            dic_inverso[valor] = duplicados[valor]  # reflejamos asi en dic_invers todas las claves que comparten el mismo valor

    # Mensaje de advertencia centralizado
    if duplicados: # solo se considera True si no es vacio
        print('Hemos encontrado valores duplicados:')
        for valor, claves in duplicados.items():
            print(f' {valor} - {claves}')

    return dic_inverso

# Ejemplo: lo substituiremos por el diccionario del cliente cuando nos lo de
enquesta = {'Ana': 7,'Dragos': 9,'Laura': 8,'Luis': 9}
resultado = invertir_diccionario(enquesta)
print('\nDiccionario Inverso:')
print(resultado)
```

##1. Comptador i endre√ßador de paraules d'un text.

El client va quedar content amb el comptador de paraules, per√≤ ara vol llegir arxius TXT i que calculi la freq√º√®ncia de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, √©s a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:


```python
import string

def contador_por_inicial(nombre_archivo):
    # cargamos el archivo
    try:
        with open(nombre_archivo, "r", encoding="utf-8") as f:
            texto = f.read().lower()  # todo min√∫sculas
    except FileNotFoundError:
        print(f"Error: No se encuentra el archivo '{nombre_archivo}'")
        return {}
    except Exception as e:
        print(f"Error al leer el archivo: {e}")
        return {}

    # limpiamos el texto de puntuaci√≥n
    for char in string.punctuation:
        texto = texto.replace(char, '')

    palabras = texto.split()  # funcion que divide texto en palabras y las pone en una lista
    contador_palabras = {}  # definimos un diccionario. Perfecto para lo que queremos. Palabra:Numero de veces que aparece la palabra

    for palabra in palabras:
        if palabra in contador_palabras:
            contador_palabras[palabra] += 1  # como la palabra ya ha aparecido antes, le sumamos el valor +1
        else:
            contador_palabras[palabra] = 1 # primera vez que aparece

    ordenado = {}  # agrupamos por letra inicial
    for palabra, freq in contador_palabras.items():
        inicial = palabra[0]  # la primera letra
        if inicial not in ordenado:
            ordenado[inicial] = {}
        ordenado[inicial][palabra] = freq

    for inicial in ordenado:  # ordenamos las palabras
        ordenado[inicial] = dict(sorted(ordenado[inicial].items()))

    ordenado = dict(sorted(ordenado.items()))  # ordenamos letras principales

    return ordenado


# subimos el archivo en Colab
from google.colab import files
print("Sube tu archivo .txt:")
uploaded = files.upload()
nombre_archivo = list(uploaded.keys())[0]  # obtenemos el nombre del archivo subido
resultado = contador_por_inicial(nombre_archivo)

if resultado:
    print("\nDiccionario final por iniciales:")
    for inicial, palabras_dict in resultado.items():
        print(f"{inicial}: {palabras_dict}")
```

##2. Conversi√≥ de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client:

```python
def convertir_lista(lista):
    convertidos = []
    no_convertidos = []

    for element in lista:
        # si el elemento es una lista o tupla, procesamos recursivamente
        if isinstance(element, (list, tuple)):
            conv, no_conv = convertir_lista(element)
            convertidos.extend(conv) # usamos .extend y no .append porque al append adicionaria la lista completa por ej [2,3] y el extend a√±ade cada elemento separado
            no_convertidos.extend(no_conv)
        else:
            # si no es lista, procesamos normalmente
            valor = str(element).strip().replace(',', '.') # str() convertimos a string, .strip elimina espacios al inicio y final y replace comas por puntos

            try: # aqui ya lo intentamos convertir a float, si falla va a no_convertidos()
                convertidos.append(float(valor))
            except ValueError:
                no_convertidos.append(element)

    return convertidos, no_convertidos


# Ejemplo
data = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]
convertidos, no_convertidos = convertir_lista(data)
print("Convertidos a float:", convertidos)
print("No convertidos:", no_convertidos)
```

##1. Generador de Contrasenyes

Explora el funcionament del m√≤dul random de la llibreria numpy. üîó Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d‚Äôaniversari o similars. Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

crear_contrasenya(10, True, True, True, True)

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).


```python
import numpy as np
import string

def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False):
    # validamos que la longitud sea positiva
    if longitud <= 0:
        print("La longitud ha de ser un n√∫mero positivo.")
        return None

    opcions = []

    # a√±adimos los tipos de caracteres seleccionados
    if majuscules:
        opcions.append(list(string.ascii_uppercase))
    if minuscules:
        opcions.append(list(string.ascii_lowercase))
    if numeros:
        opcions.append(list(string.digits))
    if signes:
        opcions.append(list("!@#$%&*?,-$"))

    # verificamos que haya al menos un tipo seleccionado
    if not opcions:
        print("Debes seleccionar al menos un tipo de car√°cter.")
        return None

    # nos aseguramos de que aparezca al menos un car√°cter de cada tipo seleccionado
    contrasenya = [np.random.choice(opcion) for opcion in opcions]

    # aplanamos opcions para tener todos los caracteres disponibles en una sola lista
    todos_caracteres = [c for opcion in opcions for c in opcion]

    # rellenamos hasta la longitud deseada con caracteres aleatorios
    if longitud > len(contrasenya):
        contrasenya += list(np.random.choice(todos_caracteres, longitud - len(contrasenya)))

    # mezclamos todo para que sea realmente aleatorio
    np.random.shuffle(contrasenya)

    # convertimos la lista en string y nos aseguramos de la longitud exacta
    contrasenya_final = ''.join(contrasenya[:longitud])

    return contrasenya_final

mi_contrasenya = crear_contrasenya(8, True, True, True, True)
print("Contrase√±a generada:", mi_contrasenya)
```

##2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. T√© un hist√≤ric de partits de futbol catal√† en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera autom√†tica, per extreure‚Äôn els resultats que necessita.

Utilitza el fitxer "historic_partits.txt"

Necessita un programa que retorni:

El nombre total de gols que ha fet cada equip.
El nom de l‚Äôequip m√©s golejador.
El nom de l‚Äôequip m√©s golejat
La classificaci√≥ global (cada vict√≤ria: 3 pts, empat 1 pts, derrota 0 pts)
Com a refer√®ncia, els equips amb m√©s punts serien els seg√ºents:

imatge resultat partits
Pots utilitzar el format que et sigui m√©s convenient.

```python
def procesar_partidos(ruta_archivo):
    equipos = {}

    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            for linea in archivo:
                linea = linea.strip()
                if not linea:
                    continue

                # Formato esperado: "Equipo1 goles1-goles2 Equipo2"
                partes = linea.split()

                # Buscamos el guion entre los goles
                pos_resultado = -1
                for i, parte in enumerate(partes):
                    if '-' in parte:
                        pos_resultado = i
                        break
                if pos_resultado == -1:
                    continue

                equipo1 = ' '.join(partes[:pos_resultado])
                resultado = partes[pos_resultado]
                equipo2 = ' '.join(partes[pos_resultado + 1:])

                try:
                    goles1, goles2 = map(int, resultado.split('-'))
                except:
                    continue

                # Inicializar si no existen
                for equipo in [equipo1, equipo2]:
                    if equipo not in equipos:
                        equipos[equipo] = {
                            'goles_favor': 0,
                            'goles_contra': 0,
                            'ganados': 0,
                            'empatados': 0,
                            'perdidos': 0,
                            'puntos': 0
                        }

                # Actualizamos goles
                equipos[equipo1]['goles_favor'] += goles1
                equipos[equipo1]['goles_contra'] += goles2
                equipos[equipo2]['goles_favor'] += goles2
                equipos[equipo2]['goles_contra'] += goles1

                # Actualizamos resultados y puntos
                if goles1 > goles2:
                    equipos[equipo1]['ganados'] += 1
                    equipos[equipo1]['puntos'] += 3
                    equipos[equipo2]['perdidos'] += 1
                elif goles1 < goles2:
                    equipos[equipo2]['ganados'] += 1
                    equipos[equipo2]['puntos'] += 3
                    equipos[equipo1]['perdidos'] += 1
                else:
                    equipos[equipo1]['empatados'] += 1
                    equipos[equipo2]['empatados'] += 1
                    equipos[equipo1]['puntos'] += 1
                    equipos[equipo2]['puntos'] += 1

        return equipos

    except FileNotFoundError:
        print("No se encontr√≥ el archivo indicado.")
        return None


def mostrar_estadisticas(equipos):
    if not equipos:
        print("No hay datos para mostrar.")
        return

    print("GOLES POR EQUIPO:")
    for equipo, datos in equipos.items():
        print(f"{equipo}: {datos['goles_favor']} goles")

    print()
    mas_goleador = max(equipos, key=lambda e: equipos[e]['goles_favor'])
    mas_goleado = max(equipos, key=lambda e: equipos[e]['goles_contra'])

    print(f"Equipo m√°s goleador: {mas_goleador} ({equipos[mas_goleador]['goles_favor']} goles)")
    print(f"Equipo m√°s goleado: {mas_goleado} ({equipos[mas_goleado]['goles_contra']} en contra)")
    print()

    print("CLASIFICACI√ìN:")
    clasificacion = sorted(equipos.items(), key=lambda x: x[1]['puntos'], reverse=True)
    for i, (equipo, d) in enumerate(clasificacion, start=1):
        print(f"{i}. {equipo} - {d['puntos']} pts (GF: {d['goles_favor']}, GC: {d['goles_contra']})")


# Programa principal
if __name__ == "__main__":
    ruta = "/content/historic partits.txt"
    equipos = procesar_partidos(ruta)
    mostrar_estadisticas(equipos)
```
